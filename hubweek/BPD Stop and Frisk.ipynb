Boston Police Department FIO (“stop and frisk”): https://data.cityofboston.gov/Public-Safety/Boston-Police-Department-FIO/xmmk-i78r/data 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
matplotlib.style.use('ggplot')
%matplotlib inline
```

```python
df = pd.read_csv("Boston_Police_Department_FIO.csv")
```

```python
df.head()
```

```python
df.describe()
```

# How many Stop and Frisks per year?

```python
df['FIO_DATE_CORRECTED'].head()
```

```python
df['NEW_FIO_DATE'] = pd.to_datetime(df['FIO_DATE_CORRECTED'])
```

```python
df['FIO_YEAR'] = df['NEW_FIO_DATE'].dt.year
```

```python
df['FIO_YEAR'] = df['FIO_YEAR'].astype('category')
```

```python
df['FIO_YEAR'].value_counts()
```

```python
df['FIO_YEAR'].value_counts().sort_index().plot(kind='bar')
plt.title("Frequency Histogram of Stop & Frisk 2011-2015")
plt.xlabel("Year")
plt.savefig("Histogram_Year.png")
```

# Freq. Histogram of various DISTS

```python
df['DIST'].value_counts().sort_index().plot(kind='bar')
```

# Freq. Histogram of Unique DIST_IDS

```python
df['DIST_ID'].value_counts().sort_index()
```

```python
df['DIST_ID'].value_counts().sort_index().plot(kind='bar')
plt.title("Frequency Histogram of S&F for Unique Districts")
plt.xlabel("Unique District Ids")
plt.savefig("Frequency Histogram of S&F for Unique Districts")
```
