```python
switched = data[data.Switched_facility.isin(['Yes', 'No'])]
switched.head()

switched["Switched_facility"] = (switched.Switched_facility == "Yes").astype("float")
```

```python
import statsmodels.api as sm
```

```python
var_columns =["Age", "House_income", "Months_pregnant", "Number_visits", "Type_facility", "Children", 
              "First_visit", "Last_visit", "Minutes_doctor", "Minutes_waiting"]
```

```python
switched_clean = switched[var_columns + ["Switched_facility"]]
```

```python
switched_clean
```

```python
import numpy as np
test_index = np.random.choice(switched_clean.index, size=0.2*len(switched_clean.index))
test = switched_clean.loc[test_index]
train = switched_clean.drop(test_index)
```

```python
logit = sm.Logit(train.Switched_facility, train[var_columns])
result = logit.fit
```
